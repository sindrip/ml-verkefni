```python
from pathlib import Path
from python_speech_features import mfcc
from python_speech_features import delta
from python_speech_features import logfbank
import scipy.io.wavfile as wav

from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

pathlist = Path('../genres').glob('**/*.wav')
flat_mfcc = []
for path in pathlist:
    print(str(path))
    (rate, sig) = wav.read(str(path))
    sig = sig[0:minsig]
    
    mfcc_feat = mfcc(sig, rate, nfft=551)
    d_mfcc_feat = delta(mfcc_feat, 2)
    dd_mfcc_feat = delta(d_mfcc_feat, 2)
    
#     print(mfcc_feat.shape)
#     print(d_mfcc_feat.shape)
#     print(dd_mfcc_feat.shape)
    test = np.hstack((mfcc_feat.flatten(), d_mfcc_feat.flatten()))
    test = np.hstack((test, dd_mfcc_feat.flatten()))
    flat_mfcc.append(test)
```

```python
ts = TSNE(n_components=2, perplexity=50.0, n_iter=300)
yfit = ts.fit_transform(flat_mfcc)
```

```python
pca = PCA(n_components=100, whiten=True)
pca.fit(flat_mfcc)
pca_data = pca.transform(flat_mfcc)
```

```python
train_size = 0.7
test_size = 0.3
new_train_mfcc = []
new_test_mfcc = []
np_new_mfcc = np.array(new_mfcc)
for i in range(10):
    rand = np.random.permutation(2000) + (2000*i)
    train = rand[:int((2000*train_size))]
    test = rand[int((2000*train_size)):int((2000*train_size)+(2000*test_size))]
    new_train_mfcc.extend(np_new_mfcc[train])
    new_test_mfcc.extend(np_new_mfcc[test])
    
print(len(new_train_mfcc))
print(len(new_test_mfcc))
new_train_mfcc_y = np.repeat(np.arange(10), (train_size*100)*20)
new_test_mfcc_y = np.repeat(np.arange(10), (test_size*100)*20)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_jobs=-1, n_estimators=500, max_features='sqrt', oob_score=True)
rf.fit(new_train_mfcc, new_train_mfcc_y)
```

```python
print("RF Out-of-bag error rate: ", 1-rf.oob_score_)
```

```python
print(rf.score(new_test_mfcc[0:20], new_test_mfcc_y[0:20]))
print(rf.predict(new_test_mfcc[200:220]))
```
